```
#Soal 1
#Graded
def letter_catalog(items, letter='A'):
  catalogs = []

  for item in items:
    if item[0] == letter:
      catalogs.append(item)

  return catalogs

#Soal 2
#Graded
def counter_item(items):
  catalogs = {}

  for item in sorted(items):
    if item in catalogs:
      catalogs[item] = catalogs.get(item) + 1
    else:
      catalogs[item] = 1

  return catalogs

#Soal 3
#Graded
fruits = ['Apple','Avocado','Banana','Blackberries','Blueberries','Cherries','Date Fruit','Grapes','Guava','Jackfruit','Kiwifruit']
prices = [6,5,3,10,12,7,14,15,8,7,9]

def fruits_prices_to_fruit_price(fruits, prices):
    zipped_fruit_price = zip(fruits, prices)
    return dict(zipped_fruit_price)

fruit_price = fruits_prices_to_fruit_price(fruits,prices)

chart = ['Blueberries','Blueberries','Grapes','Apple','Apple','Apple','Blueberries','Guava','Jackfruit','Blueberries','Jackfruit']

def total_price(dcounter, fprice):
  total = 0

  for fruit_name, fruit_count in dcounter.items():
    if fruit_name in fprice:
      total += fruit_count * fprice.get(fruit_name)

  return total

#Soal 4
#Graded

def discounted_price(total,discount,minprice=100):
    if total >= minprice:
      total -= discount / 100 * total
      return total
    else:
      return total
      
#Soal 5
#Graded

def print_summary(items,fprice):
  for fruit_name, fruit_count in counter_item(items).items():
    if fruit_name in fprice:
      print("{} {} : {}".format(fruit_count,fruit_name,fruit_count*fprice.get(fruit_name)))

  print("total : {}".format(total_price(counter_item(items),fruit_price)))
  print("discount price : {}".format(discounted_price(total_price(counter_item(items),fruit_price),10,minprice=100)))
```
