```
# >>>>> TEST >>>>>
if __name__ == '__main__':
  t1 = run()
  t2 = run()

  print('titik 1 :',t1.ambiltitik())
  print('titik 2 :',t2.ambiltitik())
  t1.tambahkan(t2)
  print('titik 2 ditambahkan ke titik 1 :',t1.ambiltitik())
```
