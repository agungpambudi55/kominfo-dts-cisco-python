```
# Soal 1
# Graded
def caesar_encript(txt,shift):
    caesar = []
 
    for i in range(len(txt)):
      if txt[i].isalpha():
        if (txt[i].islower()): 
          caesar.append(chr((ord(txt[i]) + shift - 97) % 26 + 97))
        else:
          caesar.append(chr((ord(txt[i]) + shift - 65) % 26 + 65))
      else:
        caesar.append(txt[i])
        
    return ''.join(caesar)
 
def caesar_decript(chiper,shift):
  return caesar_encript(chiper,-shift)

# Soal 2
# Graded
def shuffle_order(txt,order):
  return ''.join([txt[i] for i in order])
 
def deshuffle_order(sftxt,order):
  txt = list(range(len(sftxt)))

  for order_i in order:
    txt[order_i] = sftxt[order.index(order_i)]
   
  return ''.join(txt)

# Soal 3
# Graded
def send_batch(txt,batch_order,shift=3):
  txt_encripted = caesar_encript(txt,shift)
  if len(batch_order) > 1 and len(batch_order) < len(txt) and (len(txt) % len(batch_order)) != 0:
    txt_encripted += '_' * (len(batch_order) - (len(txt) % len(batch_order)))
  split_batch = [txt_encripted[i:i+len(batch_order)] for i in range(0, len(txt_encripted), len(batch_order))]
  batch_txt_encripted = [shuffle_order(sftxt,batch_order) for sftxt in split_batch]
  return batch_txt_encripted

def receive_batch(batch_cpr,batch_order,shift=3):
  batch_txt = [caesar_decript(deshuffle_order(i,batch_order),shift) for i in batch_cpr]
  return ''.join(batch_txt).strip('_')
```
